```python
def leiaint(msg):
    ok = False
    valor = 0
    while True:
        
        number = str(input(msg))
        if number.isnumeric():
            valor = int(number)
            ok = True
        else:
            print("\033[0;31mErro!! Digite um valor inteiro valido. \033[m")
        if ok:
            break
    return valor
entrada = leiaint("Digite um numero: !")

print(f"VocÃª digitou o valor : {entrada}")
```

```python
def leiaintnovo_int(msg):
    valor = 0
    while True:
        try:
            valor = int(input(msg))
            return valor
            
        except ValueError:
            print("\033[0;31mErro!! Digite um valor inteiro valido. \033[m")
            

  
entrada = leiaintnovo_int("Digite um valor")   
print(f" Valor digitado {entrada}")   
    
```
